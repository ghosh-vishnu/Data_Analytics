```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
sizes = [35, 33]
labels = ['Male', 'Female']
colors = ['skyblue', 'pink']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Angles for each wedge
theta = np.linspace(0, 2 * np.pi, 100)

# Draw 3D pie slices manually
start = 0
for i, s in enumerate(sizes):
    end = start + (s / sum(sizes)) * 2 * np.pi
    x = np.cos(np.linspace(start, end, 50))
    y = np.sin(np.linspace(start, end, 50))
    ax.plot_surface(
        np.outer(x, [0, 1]),
        np.outer(y, [0, 1]),
        np.outer(np.ones_like(x), [0, 0.2]),
        color=colors[i]
    )
    start = end

ax.set_title("3D Pie Chart (Matplotlib)")
plt.axis("off")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the pie chart
sizes = np.array([35, 25, 25, 15]) 
labels = ["Apples", "Bananas", "Cherries", "Dates"]
explode = [0.1, 0, 0, 0] # To slightly separate one slice

# Create the plot
fig, ax = plt.subplots()
ax.pie(sizes, 
       labels=labels, 
       autopct='%1.1f%%',  # Display percentages
       shadow=True,        # Add shadow for a 3D effect
       startangle=90,      # Start angle at 90 degrees
       explode=explode     # Explode the first slice
      )

ax.axis('equal')  # Ensures the pie chart is circular
plt.title("Fruit Sales Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = df['Salary']

fig, ax = plt.subplots(figsize=(8, 4))
counts, bins, patches = ax.hist(data, bins=20, edgecolor='white', linewidth=1)
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'orange', 'purple'] 

cmap = plt.cm.viridis_r
norm = plt.Normalize(vmin=counts.min(), vmax=counts.max())

for i, patch in enumerate(patches):
    patch.set_facecolor(cmap(norm(counts[i])))

ax.set_title('Multicolor Histogram')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
plt.show()

print(ax)
print(counts)
print(bins)
```
